```python
all_data=[]
for file in all_files:
    df = pd.read_excel(file)
    all_data.append(df)

all_data[0]
```

```python
all_data_concat = all_data_concat.drop('번호',axis='columns')
all_data_concat
```

```python
all_data_concat = all_data_concat.drop(all_data_concat.columns[[0]],axis='columns')
all_data_concat
```

#### 데이터전처리

```python
all_title = all_data_concat['제목']
all_title
```

```python
stopwords = set(stopwords.words('english'))
stopwords
```

```python
lemma = WordNetLemmatizer()
```

```python
words =[]

for title in all_title:
    EnWords = re.sub(r'[^a-zA-Z]+',' ',str(title))
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopwords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]

    words.append(EnWordsTokenStopLemma)
```

```python
print(words) # 불용어 제거하고 남은 키워드 단어
```

```python
words_reduce = list(reduce(lambda x,y : x+y,words))        #이차원 리스트를 일차원으로
print(words_reduce)
```

```python
count = Counter(words_reduce)
count
```

```python
word_count = dict()
for tag, counts in count.most_common(50):
    if len(str(tag))>1:
        word_count[tag] = counts
        print(f'{tag}:{counts}')
```
