```python
#파일병합
all_files = glob.glob('./data/myCabinetExcelData*.xls')
all_files
```

```python
all_data=[]
for file in all_files:
    df = pd.read_excel(file)
    all_data.append(df)

all_data[0]
```

```python
all_data_concat = pd.concat(all_data,axis=0,ignore_index=True)
all_data_concat
```

```python
all_data_concat.to_csv('./data/riss_bigdata.csv',index=False,encoding='utf-8')
```

```python
all_data_concat = all_data_concat.drop('번호',axis='columns')
all_data_concat
```

```python
all_data_concat = all_data_concat.drop(all_data_concat.columns[[0]],axis='columns')
all_data_concat
```

#### 데이터전처리

```python
all_title = all_data_concat['제목']
all_title
```

```python
stopwords = set(stopwords.words('english'))
stopwords
```

```python
lemma = WordNetLemmatizer()
```

```python
words =[]

for title in all_title:
    EnWords = re.sub(r'^[a-zA-Z]+',' ',str(title))
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopwords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]

    words.append(EnWordsTokenStopLemma)
```

```python
print(words) # 불용어 제거하고 남은 키워드 단어
```

```python
words_reduce = list(reduce(lambda x,y : x+y,words))        #이차원 리스트를 일차원으로
print(words_reduce)
```

```python
count = Counter(words_reduce)
count
```

```python
word_count = dict()
for tag, counts in count.most_common(30):
    if len(str(tag))>1:
        word_count[tag] = counts
        print(f'{tag}:{counts}')
```

#### 단어갯수차트

```python
sorted_keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_values = sorted(word_count.values(), reverse=True)
plt.rcParams['font.family']='NanumBarunGothic'
plt.rcParams['font.size']=13

plt.figure(figsize=(15,10))
plt.bar(range(len(word_count)),sorted_values,align='center')
plt.xticks(range(len(word_count)),list(sorted_keys),rotation=90)
plt.show()
```

#### 워드클라우드

```python
stopwords = set(STOPWORDS)
wc = WordCloud(background_color='ivory',stopwords=stopwords, width=1024, height=700)
cloud = wc.generate_from_frequencies(word_count)

plt.figure(figsize=(15,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
